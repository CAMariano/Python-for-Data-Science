# Curva de capabilidade de um gerador síncrono

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import math as m
%matplotlib inline
```

#### Limite Estatórico:
##### P=0; Q=0
#### raio = 3.V.Imáx = S;   V: tensão de saída monofásica do gerador; Imáx: corrente estatórica máxima

#### Limite Rotórico (Corrente de campo)
#### P=0 e Q=-3V^2/Xs;
#### raio= 3VEa/Xs ; Ea= tensão interna do gerador; Xs= Reatância síncrona

#### Pin=Limite devido à turbina acoplada ao eixo correspondente à potência ativa fornecida 





```python
fig = plt.figure(figsize=(5,10))
ax = fig.add_subplot(1,1,1)

#1 coluna, 1 linha, e número de plotagem 1
#subplots são úteis para mostrar múltiplos plots em uma figura.

#Dados de exercício aleatório.

V = 2300/m.sqrt(3); Xs = 1.1; Ea = 1536.55; S = 1000e3; Pin = 870.35e3*1e-6

P1 = 0; Q1 = 0; r1 = S*1e-6

P2 = 0; Q2 = (1e-6)*(-3*m.pow(V,2))/Xs; r2 = (1e-6)*(3*V*Ea/Xs)

circ1 = plt.Circle((P1,Q1), r1, fill = False)

ax.add_patch(circ1)

circ2 = plt.Circle((P2,Q2), r2, fill = False, linestyle = '--')

ax.add_patch(circ2)

plt.axvline(Pin, color='red')
plt.axhline(0, linestyle = '--', color='blue')


ax.set_xlim(0,1.5)
ax.set_ylim(-1.5,1.5)

plt.xlabel("P(MW)")
plt.ylabel("Q(MVAr)")

plt.text(0.2, 1.2, 'SUBEXCITADO')
plt.text(0.2, -1.2, 'SOBREXCITADO')

plt.show()
```
