  Merging Dataframes

```python
movies_tag = movies.merge(tags, on='movieId', how ='inner')
movies_tag.head()
```

```python
del movies_tag['timestamp']
movies_tag.head()
```

```python
avg_ratings = ratings.groupby('movieId', as_index=False).mean()

del avg_ratings['userId']
del avg_ratings['timestamp']
```

```python
box_office = movies.merge(avg_ratings, on='movieId', how='inner')
box_office.head(20)
```

```python
highly_rated = box_office['rating']>=4.0

box_office[highly_rated]
```

```python
low_rated = box_office['rating']<=2.0

box_office[low_rated]
```

```python
is_SciFi = box_office['genres'].str.contains('Sci-Fi')
```

```python
box_office[highly_rated & is_SciFi]
```
